##  Abstract

This study was conducted to determine if the price per square foot of a single family home/dwelling has changed significantly with the presence of COVID-19. The creation of a  pricing tool was used to answer that question. With the aim of future pricing predictability the tool was used to learn from previous Real estate transactions spanning from January 2, 2018 to December 31, 2020. The tool was meant to navigate/predict new prices per square foot of the market with the presence of COVID-19.There were 10,584 qualifying real estate transactions used for learning. All the transactions used for learning were properties that were included in building class categories that a single familes could occupy. The results show that the average price per square foot of single family home/dwellings has decreased significantly in New York City.

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 1000, stop = 10000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
def plot_distribution(inp):
    plt.figure(figsize = (10,10))
    ax = sns.distplot(inp)
    plt.axvline(np.mean(inp), color ="k", linestyle="dashed",linewidth = 2)
    _, max_ = plt.ylim()
    plt.text( inp.mean() + inp.mean()/10,
            max_ - max_ / 10,
            "Mean: {:2f}".format(inp.mean()),)
    return plt.figure
```

```python
def compare_2_groups(arr_1,arr_2, alpha, sample_size):
    stat, p = ttest_ind(arr_1,arr_2)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    if p > alpha:
        print('Same distributions (fail to reject H0)')
    else:
        print('Different distributions (reject H0)')
```
